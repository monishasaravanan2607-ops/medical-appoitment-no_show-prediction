```python
import pandas as pd
df = pd.read_csv("/Users/kaviraj/Desktop/GUVI/Project3/regression/r_feature_engineering.csv")
df
```

# Target variable analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['daily_appointments'], kde=True)
plt.title("Distribution of Daily Appointments")
plt.show()
```

The target variable shows significant variation across days. Some days have very low appointments while a few days experience exceptionally high appointment counts, indicating possible demand spikes.

# Outlier Analysis

```python
sns.boxplot(x=df['daily_appointments'])
plt.show()
```

# Time Trend Analysis

```python
plt.figure(figsize=(12,5))
plt.plot(df['appointment_date_continuous'],
         df['daily_appointments'])
plt.title("Daily Appointments Over Time")
plt.show()
```

# Month-wise Analysis

```python
sns.boxplot(x='month',
            y='daily_appointments',
            data=df)
plt.show()
```

# Weekend vs Weekday Analysis

```python
sns.barplot(
    x='is_weekend',
    y='daily_appointments',
    data=df
)
plt.show()
```

# Correlation Heatmap

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr,
            cmap='coolwarm')
plt.show()
```

# Lag Feature Analysis

```python
sns.scatterplot(
    x='lag_1',
    y='daily_appointments',
    data=df
)
plt.show()
```

# Rolling Mean Analysis

```python
sns.scatterplot(
    x='rolling_mean_7',
    y='daily_appointments',
    data=df
)
plt.show()
```
