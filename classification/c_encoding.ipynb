```python
import pandas as pd

df = pd.read_csv("/Users/kaviraj/Desktop/GUVI/Project3/classification/c_featured_data.csv")

df
```

```python
print(df.columns.tolist())
```

```python
df.drop(['place','appointment_date_continuous'], axis=1, inplace=True)
```

# 1. Identify Column types

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include='object').columns

print("numerical_cols:", numerical_cols)
print("categorical_cols:", categorical_cols)
```

```python
for col in ['specialty', 'gender', 'no_show', 'disability', 'place',
            'appointment_shift', 'rain_intensity', 'heat_intensity', 'day_of_week']:
    print(col, df[col].nunique())
    print(col, df[col].unique())
```

# 2. Encoding

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
# Create a LabelEncoder instance
le = LabelEncoder()

# Apply Label Encoding to the 'Gender' column
df['no_show'] = le.fit_transform(df['no_show'])

# Display the DataFrame with the new encoded column
display(df)
```

# One-hot Encoder

```python
onehot_cols = ['specialty', 'gender', 'disability', 'appointment_shift', 'day_of_week']

# Apply One-Hot Encoding
df = pd.get_dummies(df, columns = onehot_cols, drop_first = True)

df.head(5)
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(
    categories=[
        ['no_rain', 'weak', 'moderate', 'heavy'],  # rain_intensity
        ['heavy_cold', 'cold', 'mild', 'warm', 'heavy_warm']  # heat_intensity
    ]
)

df[['rain_intensity', 'heat_intensity']] = oe.fit_transform(
    df[['rain_intensity', 'heat_intensity']]
)

df.head(5)
```
