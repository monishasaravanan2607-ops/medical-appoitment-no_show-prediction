```python
# Missing percentage

df.isnull().sum()/len(df)*100
```

```python
# Handling missing values for categorial data
# speciality, disability, place
# Unknown

df['specialty'] = df['specialty'].fillna('Unknown')
df['disability'] = df['disability'].fillna('Unknown')
df['place'] = df['place'].fillna('Unknown')
```

```python
# Handling missing values for numerical data
# age, average_temp_day, average_rain_day, max_temp_day, max_rain_day
# median

missing_cols = ['age', 'average_temp_day', 'average_rain_day', 'max_temp_day', 'max_rain_day']

for col in missing_cols:
    df[col] = df[col].fillna(df[col].median())
```
